```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('df_cust_financial_with_cluster.csv')
df = df.drop(columns=["id","join_key",
                      "join_key_2",'new_col'])
df
```

```python
import pycaret
from pycaret.classification import *

reg2 = setup(data = df, target = 'clusters',train_size=0.8, session_id=123, use_gpu=True)
```

```python
models()
```

```python
# clf = create_model('xgboost')
```

```python
best_classification_model = compare_models(exclude = ['ada','lr','gbc','svm'],sort='Recall', n_select=3)
```

```python
# plot_model(clf, plot='feature')
```

```python
tuned_model=[tune_model(i) for i in best_classification_model]
```

```python
blended_cls=blend_models(estimator_list=tuned_model,fold=4,optimize='Recall')
```

```python
final_model_cls = finalize_model(blended_cls)
```

```python
test_df = pd.read_csv('df_cust_non_financial.csv')
prediction=predict_model(final_model_cls,data=test_df)
```
